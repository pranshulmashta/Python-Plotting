```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import style
style.use('ggplot')
```

```python
years= gas.Year[::3]
years
```

```python
plt.figure(figsize=(20,10))
plt.title("Gas Prices in USD")
#plt.plot(gas.Year, gas.UK, label='UK', marker='.')
#plt.plot(gas.Year, gas.Canada, label= 'Canada')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='*')
                               
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.legend()
plt.xticks(years)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title("Gas Prices in USD", fontdict={'fontweight':'bold', 'fontsize': 18})
plt.plot(gas.Year, gas.UK, label='UK', marker='.',)
plt.plot(gas.Year, gas.Canada, label= 'Canada',marker='*')
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.legend()
plt.xticks(years)
plt.savefig('sample.png', dpi=250)

plt.show()
```
